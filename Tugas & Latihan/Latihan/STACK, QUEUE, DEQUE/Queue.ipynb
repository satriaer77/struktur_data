```python
import sys
sys.path.append("..")

from modules import queueV2 as que

def tampilkanKar(lst,typ) :
    if typ == "t" :
        for l in lst :
            print(l,end="\t")
    elif typ == "s":
        kar = """ """
        for l in lst :
            kar+="""%s"""%(l)
        return kar
        
def dragonSnake(players,count) :
    queuePlayers = que.createQueue()
    
    for player in players :
        que.enqueue(queuePlayers,player)
    print("""
    +---------Pemain---------+
    
               %s
    
    +------------------------+
    
    """%tampilkanKar(queuePlayers,"s"))
    
    while que.size(queuePlayers) > 1 :
        for i in range(count) :
            play = que.enqueue(queuePlayers,que.dequeue(queuePlayers))
            tampilkanKar(play,"t")
            print("====Check====")
        que.dequeue(queuePlayers)
    
    return """
    ===== Pemenang =====
      
       %s
          
    ====================
    
    """% (queuePlayers[0])

print(dragonSnake(["""
   Ramon
  /------\\
 | 0  0 |
 \\------/
    |
   /|\\
  / | \\
    |  
   / \\
""",
"""
   Jaka
    /\\
 /------\\
 | 0 | 0 |
 \\------/
    |
   /|\\
  / | \\
    |  
   / \\
""",
"""
   Jessica
  ""/\\""
 /------\\
 | 0 | 0 |
 \\------/
    |
   /|\\
  / | \\
    |  
   / \\
"""],3))
    
```

```python
import sys
sys.path.append("..")

from modules import queueV2 as que
tasks = [5,4,3]

def schTask(tasks,limit) :
    taskQueue = que.createQueue()
    total     = 0
    for task in tasks :
        que.enqueue(taskQueue,task)
    
    while que.size(taskQueue) > 1 :
        for i in range(limit) :
            if que.size(taskQueue) > 1 :
                time = que.dequeue(taskQueue)-limit
                total+=limit
                if time <= 0 :
                    que.dequeue(taskQueue)
                que.enqueue(taskQueue,time)
                print(time)
                print(taskQueue)
            
    return total

print(schTask(tasks,3))

```
